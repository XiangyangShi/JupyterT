```python
#minimax
count = 0
def minimax(root,depth):
    global count
    #print depth,root.side
    if depth == 0:
        #print "come to a end"
        count +=1
        return root.evaluate("Star" if Firsthand else "Circle")
    if (root.hasoneleft("Star"))==False or root.hasoneleft("Circle")==False:
        #print "none left"
        count +=1
        return root.evaluate("Star" if Firsthand else "Circle")
    root.expand()
#     print "depth",depth
#     for i in root.child:
#         i.show()
    
    if root.movename=="pass" and root.child[0].movename=="pass":
        #print "double pass"
        count +=2
        root.movename += " pass"
        return root.evaluate("Star" if Firsthand else "Circle")
    valuelist = []
    for i in root.child:
        value = minimax(i,depth-1)
        valuelist.append(value)
    if root.side == ("Star" if Firsthand else "Circle"):
        #print "max"
        count +=1
        maxvalue = max(valuelist)
        maxindex = valuelist.index(maxvalue)
        root.movename += " "+root.child[maxindex].movename
        return maxvalue
    else:
        #print "min"
        count +=1
        minvalue = min(valuelist)
        minindex = valuelist.index(minvalue)
        root.movename += " "+root.child[minindex].movename
        return minvalue
```

```python
#alphabeta
count = 0
def alphabeta(root,depth):
    global count

    if depth == 0:
#         print "leaf ",root.evaluate()
        count +=1
        return root.evaluate("Star" if Firsthand else "Circle")
    if (root.hasoneleft("Star"))==False or root.hasoneleft("Circle")==False:
#         print "eat all"
        count +=1
        return root.evaluate("Star" if Firsthand else "Circle")
    root.expand()
    
    if root.movename=="pass" and root.child[0].movename=="pass":
#         print "doubl pass"
        root.movename += " pass"
        count +=2
        return root.evaluate("Star" if Firsthand else "Circle")
    
    for i in root.child:
        i.alpha = root.alpha
        i.beta = root.beta
        
        value = alphabeta(i,depth-1)
        
        if root.side == ("Star" if Firsthand else "Circle"):
            # print "max"
            if value > root.value:
                root.value = value
                root.movename = root.movename.split()[0] +" "+ i.movename
            if root.value >= root.beta:
                count +=1
                return root.value
            if root.value > root.alpha:
                root.alpha = root.value
            
        else:
            # print "min"
            if value < root.value:
                root.value = value
                root.movename = root.movename.split()[0] +" "+ i.movename
            if root.value <= root.alpha:
                count += 1
#                 print "short cut on min"
#                 print "rootalpha",root.alpha
#                 print "rootvalue",root.value
                return root.value
            if root.value < root.beta:
                root.beta = root.value
    count += 1
    return root.value      
```
