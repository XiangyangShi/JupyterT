```python
def maxvalue(index,Tree,nodes):
    if(Tree[index][3]!='N'):
        parentindex = int(Tree[index][3])
    else:
        parentindex=0
    if(Tree[index][2]=='N'):
        nodes=nodes+1
        Tree[parentindex][1] = str(min(int(Tree[index][1]), int(Tree[parentindex][1])))
        return str(Tree[parentindex][1])
    else:
        nodes=nodes+1
        children=Tree[index][2].split(',')
        for i in range(len(children)):
            value=minvalue(int(children[i]),Tree,nodes)
            if (int(Tree[index][1]) <= value):
                continue
            else:
                Tree[parentindex][1] = value
        return Tree[parentindex][1]

```

```python
def minvalue(index,Tree,nodes):
    if (Tree[index][3] != 'N'):
        parentindex = int(Tree[index][3])
    else:
        parentindex = 0
    if(Tree[index][2]=='N'):
        nodes=nodes+1
        Tree[parentindex][1]=str(max(int(Tree[index][1]),int(Tree[parentindex][1])))
        return Tree[parentindex][1]
    else:
        nodes=nodes+1
        children=Tree[index][2].split(',')
        for i in range(len(children)):
            value=maxvalue(int(children[i]),Tree,nodes)
            if(int(Tree[index][1])>=value):
                continue
            else:
                Tree[index][1]=value
        return Tree[index][1]
```
